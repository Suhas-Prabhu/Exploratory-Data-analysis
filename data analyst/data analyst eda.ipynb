```python
# check the different jobs 
data['Job Title'].value_counts()
```

```python
data['Job Title'].value_counts()[:20]
```

We can modify the above data to classify senior and junior properly


```python
data["Job Title"] = data["Job Title"].str.replace('Sr. Data Analyst','Senior Data Analyst')
data["Job Title"] = data["Job Title"].str.replace('Sr Data Analyst','Senior Data Analyst')
data["Job Title"] = data["Job Title"].str.replace('Data Analyst Senior','Senior Data Analyst')
data["Job Title"] = data["Job Title"].str.replace('Jr. Data Analyst','Junior Data Analyst')
data["Job Title"] = data["Job Title"].str.replace('Jr Data Analyst','Junior Data Analyst')
data["Job Title"] = data["Job Title"].str.replace('Data Analyst Junior','Junior Data Analyst')
data["Job Title"].value_counts()
```

```python
data["Job Title"].isnull().sum() # no missing values found in the j ob title
```

Lets move on to the Salary column

```python
data["Salary Estimate"].isnull().sum()
```

As we can see that there is one single missing value in salary.Its a very important column for analysis and hence we cannot ignore the missing value. Even if you have no domain experience we can fill in the missing value by checking at similar jobs from the same company. 

Still filling the value is a complex task as it contains text, numberic and $ value. so lets see how we can proceed

```python
#explore the missing value row
data[data['Salary Estimate'].isnull()]
```

```python
""" The company name is Protingent\n4.4 and job title is Data management analyst |||. lets filter these values"""
data[data['Company Name']=='Protingent\n4.4']
```

```python
data[data['Job Title']=='Configuration/Data Management Analyst III']
```

We can observe that there only one  requirement for that job and there are 2 job requirements from the same company. if the job description are similar then we can set the same pricing

```python
print(data['Job Description'][2123])
data['Job Description'][2149]
```

Even though thereare similarities i guess we can drop the row with the missing value. 

Furthermore as it is a salary estimate column we can further get more data from it. We can calculate Max salary, Min salary and average salary.

```python
data.drop(2149, inplace=True)
```

```python
data['max salary'] = data['Salary Estimate'].str[6:10].str.replace('K','').str.lstrip('$').str.strip().astype("float")
data['max salary']
```

```python
data["min salary"] = data['Salary Estimate'].str.lstrip('$').str[:3].str.replace('K','').str.strip().astype('float')
data['min salary']
```

```python
data["avg salary"] = (data["min salary"]+ data['max salary'])/2
data["avg salary"]
```

```python
data[['Salary Estimate','min salary','max salary','avg salary']].sample(10)
```

Moving on to job description, we can extract valuable information from this column. Here we can extract different language requirements for the job.

```python
data["Job Description"].isnull().sum()
```

```python
data['python job'] = data["Job Description"].str.contains('python', na =False, case =False)
data["python job"].value_counts()
```

```python
data['java job'] = data["Job Description"].str.contains('java', na =False, case =False)
data['java job'].value_counts()
```

```python
data['sql job'] = data["Job Description"].str.contains('sql', na =False, case =False)
data['sql job'].value_counts()
```

```python
data['tableau job'] = data["Job Description"].str.contains('tabeau', na =False, case =False)
data['tableau job'].value_counts()
```

```python
data['excel job'] = data["Job Description"].str.contains('excel', na =False, case =False)
data['excel job'].value_counts()
```

```python
data['HTML job'] = data["Job Description"].str.contains('HTML', na =False, case =False)
data['HTML job'].value_counts()
```

```python
# moving on to rating
print(data['Rating'].isnull().sum())
# we will keep the rows which has missing value on Rating, we ant to use salary and job title info on these rows.
data["Company Name"].sample(10)
```

##### Lets start visualizing the data to get more insights on the data

```python
salary_min = px.histogram(analysis_data,
                             x= 'min salary',
                             title='Minimum Salary distribution of Data Analyst Jobs',
                             marginal="box",
                             hover_data = analysis_data[['Job Title', 'Company Name']])
salary_max = px.histogram(analysis_data,
                             x= 'max salary',
                             title='Maximum Salary distribution of Data Analyst Jobs',
                             marginal="box",
                             hover_data = analysis_data[['Job Title', 'Company Name']])
salary_avg = px.histogram(analysis_data,
                             x= 'avg salary',
                             title='Average Salary distribution of Data Analyst Jobs',
                             marginal="box",
                             hover_data = analysis_data[['Job Title', 'Company Name']])
```

```python
salary_min.show()
salary_max.show()
salary_avg.show()
```

### Salary visualization

- As expected we have a skewed distribution and a few outliers. 
- The majority of the minimum salary count lies between 40k-65k and maximum salary count lies between 65K-105k.
- On average The salaries lie in between 55k-85k

```python
# visualizing the top 20 jobs with number of available spots
jobs_figure = go.Figure()
jobs_figure.add_trace(go.Scatter(
  x = analysis_data['Job Title'].value_counts()[:20].index,
  y = analysis_data['Job Title'].value_counts()[:20].values,
  name = "Number of Job openings ",
  mode ='markers+text+lines',
  marker_color ="black",
  marker_size = 10,
  text = analysis_data['Job Title'].value_counts()[:20].values,
  textposition ='top center',
  line=dict(color='red',dash='dash'),
))
jobs_figure.update_layout(
  title ="<b>number of job openings by job titles</b>",
  xaxis_title ="<b>job titles</b>",
  yaxis_title ="<b>Job Openings</b>",
  template ="seaborn",
  font =dict(
    size =12,
    color ='Black',
    family ='Oswald, sans-serif',
  ),
  xaxis=dict(showgrid=True),
  yaxis=dict(showgrid=True),
  yaxis2=dict(showgrid=True,overlaying='y',side='right',title='<b>Number of Job Openings</b>'),
  legend=dict(yanchor="top",
              y=1.3,
              xanchor="left",
              x=0.78)
  )
```
